## Iterate through all images in batch

```python
def dataload(loader,m):    
    data_iter = iter(loader)
    images, labels = next(data_iter)
#     print('initial image size = ', images.shape)
    images = images.reshape(m,-1).t() /255
#     print('flattened image size = ', images.shape)
    (nx,m) = images.shape # m is no of training exmples 
    return images,labels,nx,m
```

## Helper functions





$sigmoid( w^T x + b) = \frac{1}{1 + e^{-(w^T x + b)}}$



```python
def sigmoid(x):
    return (1 / (1+torch.exp(-x)))
```

## Forward Propagation:

- Get X
- Compute $A = \sigma(w^T X + b) = (a^{(1)}, a^{(2)}, ..., a^{(m-1)}, a^{(m)})$
- Calculate the cost function: $J = -\frac{1}{m}\sum_{i=1}^{m}y^{(i)}\log(a^{(i)})+(1-y^{(i)})\log(1-a^{(i)})$

Formulas:

$$ \frac{\partial J}{\partial w} = \frac{1}{m}X(A-Y)^T\tag{7}$$
$$ \frac{\partial J}{\partial b} = \frac{1}{m} \sum_{i=1}^m (a^{(i)}-y^{(i)})\tag{8}$$

```python
def parameters(w, b, X, Y,m):
    A = sigmoid(torch.mm(w.t(), X)+b) # dimension (1 x m)
    cost = -(torch.mm(Y , torch.log(A.t())) + torch.mm((1-Y) , torch.log( (1-A).t() ) )) / m
    dz = A - Y
    db = (torch.sum(dz))  /  m
    dw = torch.mm(X, dz.t())
    
    assert(dw.shape == w.shape)
    
    return cost,dw,db
```

## Predicition 

```python
def predict(w,b,X):
    A = sigmoid(torch.mm(w.t(), X)+b)
    
    Y_prediction = torch.zeros((1,X.shape[1])).cuda()
    for i in range(X.shape[1]):
        
        if A[0,i] > 0.5:
            Y_prediction[0,i] = 1
        else:
            Y_prediction[0,i] = 0
            
    return Y_prediction
```

## Training the model

Given an initial value of weights, biases, iteration and learning rate, the below block of code will try to
optimize the values of weight and biases such that we get a higher training accuracy.

```python
def optimize(w, b, X, X_test, Y, Y_test,num_iterations, alpha ,m, vdw, vdb, beta, sdw, sdb, beta2, itr):
    
    cost_arr = []
    Y_p_train = []
    Y_p_test = []
    for i in range(num_iterations):
        
        cost,dw,db = parameters(w, b, X, Y,m)
        cost_arr.append(cost)
        
        vdw = beta * vdw + (1-beta) * dw
        vdb = beta * vdb + (1-beta) * db
       
        
        w = w - alpha * vdw
        b = b - alpha * vdb
        
        
        
        if i % 100 == 0:
            
            Y_prediction_train = predict(w, b, X)
            t1 = 100 - torch.mean(torch.abs(Y_prediction_train - Y)) * 100
            Y_p_train.append(t1)
            
            Y_prediction_test = predict(w, b, X_test)
            t2 = 100 - torch.mean(torch.abs(Y_prediction_test - Y_test)) * 100
            Y_p_test.append(t2)
            
    print("test accuracy: ", float(t2), "train accuracy: ",float(t1), "cost: ", float(cost))
    
    return cost_arr, w, b, Y_p_train, Y_p_test
```

# Optimazation of weights and biases

```python
itr = 1300 // bs
ca = []
yptr = []
ypts = []
epochs = 100         # hyper-parameters
alpha = 0.1
beta = 0.9
beta2 = 0.999
decay_rate = 0

for i in range(1,epochs):
    
    data_iter = iter(trainloader)
    
    alpha = alpha / (1+ decay_rate*i)  # learning rate decay
    print("\n epoch: ", i,'\t alpha = ',alpha)
    
    for j in range(itr):

        images, labels = next(data_iter)
        x = images.shape[0] 
        images = images.reshape(x,-1).t() /255
        X = images.cuda() # (nx x m)
        Y = labels.cuda() 
        Y = torch.reshape(Y,(1,x)) # (1 x m)
        Y = Y.float().cuda()
        
        images_test,labels_test,nx,n = dataload(testloader,90)
        X_test = images_test.cuda() # (nx x m)
        Y_test = labels_test.cuda() 
        Y_test = torch.reshape(Y_test,(1,90)) # (1 x m)
        Y_test = Y_test.float().cuda()
        
        
        cost_arr, w, b, Y_p_train, Y_p_test = optimize(w, b, X, X_test, Y,Y_test,1, alpha, x,vdw,vdb,beta,
                                                      sdw, sdb, beta2, itr)
        ca = ca + cost_arr
        yptr = yptr + Y_p_train
        ypts = ypts + Y_p_test
```

# Calculation of Test and Train Accuracy

```python
plt.ylabel('Accuracy(in %)')
plt.xlabel('no. of iteration')
plt.plot(yptr, label = 'Train Accuracy')
plt.plot(ypts, label ='Test Accuracy')
plt.legend()
```

```python
plt.ylabel('Cost')
plt.xlabel('no. of iteration')
plt.plot(ca, label ='Cost')
plt.legend()

print("Train Accuracy: ", float(Y_p_train[-1]), "\t Test Accuracy: ", float(Y_p_test[-1]))
```
