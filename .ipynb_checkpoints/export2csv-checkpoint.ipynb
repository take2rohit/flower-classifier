```python
import torch
from PIL import Image
import numpy as np
import os
import pandas as pd
import matplotlib.pyplot as plt
from time import sleep 
```

# Creation of csv files using images

1. Changed directory to folder which contains images of rose
2. Stored image in an array
3. Resized it
4. Converted to numpy array
5. Flattened the image to a row vector and normalized it 
6. Reshaped to ensure that its not a rank 1 matrix
7. Vertically stacked all the vectors
8. Horizontally stacked ones coloum vector at the end. (Here 1 represents rose)
7. Stored each of them in a seperate csv file

Dataset link: https://www.kaggle.com/alxmamaev/flowers-recognition/downloads/flowers.zip/2

```python
path = '/home/rohit/Desktop/flower/train' # Path for roses
os.chdir(path)

data = []
for pic in os.listdir(path):
    if pic.endswith("r.jpg"):
        im = Image.open(pic)
        im = im.resize((150,100),Image.ANTIALIAS)
        im = np.array(im)
        im = im.ravel() / 255
        im = np.append(im,[1])
        im = np.reshape(im,(1,im.size))
        data.append(im)
    elif pic.endswith("s.jpg"):
        
        
data = np.vstack(data)
data = np.hstack((data,np.ones((np.shape(data)[0],1)))) # rose = 1

dfr = pd.DataFrame(data)
np.shape(data)
```

```python
dfr.to_csv('rose.csv',  header = False, index = False)
```

```python
path = '/home/rohit/Desktop/flowers/sunflower' # Path for sunflower
os.chdir(path)

data = []
for pic in os.listdir(path):
    if pic.endswith(".jpg"):
        im = Image.open(pic)
        im = im.resize((150,100),Image.ANTIALIAS)
        im = np.array(im)
        im = im.ravel() / 255
        im = np.reshape(im,(1,im.size))
        data.append(im)
        
data = np.vstack(data)
data = np.hstack((data,np.zeros((np.shape(data)[0],1)))) # sunflower = 0

dfs = pd.DataFrame(data)
dfs.head()
```

```python
dfs.to_csv('sunflower.csv',  header = False, index = False)
```
